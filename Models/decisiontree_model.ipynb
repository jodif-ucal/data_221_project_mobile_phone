```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from data_preprocessing import features_train, features_test, labels_train, labels_test
```

```python
# Decision Tree model
model = DecisionTreeClassifier(random_state=42)
model.fit(features_train, labels_train)

# Predictions
y_prediction = model.predict(features_test)
# Accuracy
accuracy = accuracy_score(labels_test, y_prediction)
accuracy_percentage = round(accuracy * 100, 2)
print("Model Accuracy is:", accuracy_percentage, "%")

# Classification report
print(classification_report(labels_test, y_prediction))
```

```python
confusedMatrix = confusion_matrix(labels_test, y_prediction)
plt.figure(figsize=(8, 6))
plt.imshow(confusedMatrix, cmap="Blues")
plt.colorbar()

plt.xticks([0, 1, 2, 3], ["Low", "Medium", "High", "Very High"])
plt.yticks([0, 1, 2, 3], ["Low", "Medium", "High", "Very High"])

# Add numbers inside boxes
for i in range(len(confusedMatrix)):
    for j in range(len(confusedMatrix[i])):
        plt.text(j, i, confusedMatrix[i][j], ha="center", va="center")

plt.title("Confusion Matrix of price classifications by the Decision Tree Model")
plt.xlabel("Predicted phone prices")
plt.ylabel("True phone prices")

plt.show()
```

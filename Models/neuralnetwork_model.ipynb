```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    ConfusionMatrixDisplay
)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, InputLayer
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, roc_auc_score, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Import the shared preprocessing variables from group script
from data_preprocessing import features_train, labels_train, features_test, labels_test, feature_matrix

# Set random seed as taught in the slides to ensure reproducibility
tf.random.set_seed(1)

# Encode the outcome into binary values for the neural network
one_hot_labels_train = to_categorical(labels_train)
one_hot_labels_test = to_categorical(labels_test)
```

```python
# Define the model as a sequence of layers
neural_network_model = Sequential()

neural_network_model.add(InputLayer(shape=(20,)))

# Hidden Layers
neural_network_model.add(Dense(16, activation='relu'))
neural_network_model.add(Dense(8, activation='relu'))

# Output Layer
neural_network_model.add(Dense(4, activation='softmax'))

# Compile
neural_network_model.compile(loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model for 50 iterations (epochs)
# We use the one-hot encoded labels (0-3) created in the first cell
print("Training the Neural Network...")
history = neural_network_model.fit(features_train, one_hot_labels_train, epochs=50, verbose=1)

# Evaluate the final performance on the testing set (the 25% split)
model_performance = neural_network_model.evaluate(features_test, one_hot_labels_test)

print(f"\nOverall testing loss: {model_performance[0]:.4f}")
print(f"Overall testing accuracy: {model_performance[1]:.4f}")
```

```python
# 1. Convert probability outputs to class labels (0, 1, 2, or 3)
# np.argmax picks the index of the highest probability in each row
y_test_pred_prob = neural_network_model.predict(features_test)
y_test_pred = np.argmax(y_test_pred_prob, axis=1)

# 2. Detailed Metrics Loop
for class_value in range(0, 4):
    predicted_labels = []
    actual_labels = []

    for index in range(len(y_test_pred)):
        # Binary check for the current class
        if y_test_pred[index] == class_value:
            predicted_labels.append(1)
        else:
            predicted_labels.append(0)

        if labels_test.iloc[index] == class_value:
            actual_labels.append(1)
        else:
            actual_labels.append(0)

    # Calculate metrics
    accuracy = accuracy_score(actual_labels, predicted_labels)
    precision = precision_score(actual_labels, predicted_labels, zero_division=0)
    recall = recall_score(actual_labels, predicted_labels, zero_division=0)
    f1 = f1_score(actual_labels, predicted_labels, zero_division=0)
    roc_auc = roc_auc_score(actual_labels, predicted_labels)

    print(f"Accuracy score for class label {class_value}: {accuracy:.2f}")
    print(f"Precision score for class label {class_value}: {precision:.2f}")
    print(f"Recall score for class label {class_value}: {recall:.2f}")
    print(f"F1 score for class label {class_value}: {f1:.2f}")
    print(f"ROC-AUC score for class label {class_value}: {roc_auc:.2f}\n")
```

```python
# 1. Create the Confusion Matrix
conf_matrix = confusion_matrix(labels_test, y_test_pred)

# 2. Display the Matrix
display = ConfusionMatrixDisplay(confusion_matrix=conf_matrix,
                                 display_labels=["Low", "Medium", "High", "Very High"])

# Using the 'Blues' color map to match the KNN and Random Forest notebooks
display.plot(cmap=plt.cm.Blues)

# Formatting titles and labels
plt.title("Confusion Matrix of Price Classifications by the Neural Network")
plt.xlabel("Predicted Phone Prices")
plt.ylabel("True Phone Prices")
plt.show()
```

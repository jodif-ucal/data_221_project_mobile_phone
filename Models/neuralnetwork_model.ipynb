```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, InputLayer
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, roc_auc_score, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Import the shared preprocessing variables from group script
from data_preprocessing import features_train, labels_train, features_test, labels_test, feature_matrix

# Set random seed as taught in the slides to ensure reproducibility
tf.random.set_seed(1)

# Encode the outcome into binary values for the neural network
one_hot_labels_train = to_categorical(labels_train)
one_hot_labels_test = to_categorical(labels_test)
```

```python
# Define the model as a sequence of layers
neural_network_model = Sequential()

neural_network_model.add(InputLayer(shape=(20,)))

# Hidden Layers
neural_network_model.add(Dense(16, activation='relu'))
neural_network_model.add(Dense(8, activation='relu'))

# Output Layer
neural_network_model.add(Dense(4, activation='softmax'))

# Compile
neural_network_model.compile(loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model for 50 iterations (epochs)
# We use the one-hot encoded labels (0-3) created in the first cell
print("Training the Neural Network...")
history = neural_network_model.fit(features_train, one_hot_labels_train, epochs=50, verbose=1)

# Evaluate the final performance on the testing set (the 25% split)
model_performance = neural_network_model.evaluate(features_test, one_hot_labels_test)

print(f"\nOverall testing loss: {model_performance[0]:.4f}")
print(f"Overall testing accuracy: {model_performance[1]:.4f}")
```
